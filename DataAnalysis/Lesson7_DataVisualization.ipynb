# Sample Superstore

```python
df = pd.read_excel("./Data Visualization/Sample - Superstore.xls", sheet_name="Orders")
df.head()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df.head()
```

```python
df["Year"] = df["Order Date"].dt.year
```

```python
df.head()
```

```python
fig, ax1 = plt.subplots(figsize = (10, 4))
sum_df = df.groupby("Year").Sales.sum()

ax1.bar(sum_df.index, sum_df.values, color = "#ff94e0", width = 0.3) #tạo bar chart

ax1.spines[:].set_visible(False) # bỏ khung tứ phía

ax1.set_xticks([2018, 2019, 2020, 2021]) # set các giá trị trên trục x

ax1.set_yticklabels([str(int(x / 1000)) + "$" for x in ax1.get_yticks()]) # set label cho các giá trị trên trục y

# điền số lên trên đầu mỗi bar chart
for p in ax1.patches:
    w, h = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax1.annotate("{:.2f}".format(float(h/1000)), (x, h + 20000)) #.annotate(text, xy)
    #{:.2f}".format(a)  -> 2 digits after comma

ax1.set_ylabel("Revenue (Thousand)") # set label trục oy
ax1.set_xlabel("Year") # set label trục ox

profit_df = df.groupby("Year").Profit.sum()
ax2 = ax1.twinx()
ax2.plot(profit_df.index, profit_df.values, color = "grey", linewidth = 2, marker = "*")
ax2.spines[:].set_visible(False) # bỏ khung tứ phía

ax2.set_xticks([2018, 2019, 2020, 2021]) # set các giá trị trên trục x

ax2.set_yticklabels([str(int(x / 1000)) + "$" for x in ax1.get_yticks()]) # set label cho các giá trị trên trục y

ax2.set_ylabel("Profit (Thousand)") # set label trục oy
ax2.set_xlabel("Year") # set label trục ox

ax1.tick_params(left = False, bottom = False) # bỏ đầu nến
ax2.tick_params(right = False, bottom = False) # bỏ đầu nến
```

```python
fig, ax3 = plt.subplots(figsize = (10, 4))

ax3.plot(df3["Month"], df3["Processed"], color = "b") #tạo line chart
ax3.plot(df3["Month"], df3["Received"], color = "grey")

ax3.spines[["top", "right"]].set_visible(False) # bỏ khung 

ax3.set_ylabel("Number of tickets")

lst = [ax3.lines[0], ax3.lines[1]]
for line in lst:    
    for x_value, y_value in zip(line.get_xdata()[7:], line.get_ydata()[7:]): 
        ax3.annotate('{}'.format(y_value),(x_value,y_value+10))
        if line == ax3.lines[0]:
            ax3.scatter(x_value, y_value, color = "blue")
        else:
            ax3.scatter(x_value, y_value, color = "grey")


ax3.text(x = 11.5, y = 170, s = "Received", size = 10, weight = "bold", color = "grey")
ax3.text(x = 11.5, y = 130, s = "Processed", size = 10, weight = "bold", color = "b")


ax3.axvline(x = 4, color = "grey")
ax3.set_yticks([i for i in range(0, 350, 50)]) # set các giá trị trên trục y


ax3.text(x=-0.5, y=340, s='Please approve the hire of 2 FTEs', size=17, weight='bold')
ax3.text(x=-0.5, y=320, s='to backfill those who quit in the past year', size=12)
ax3.text(x = 4, y = 260, s=''' 2 employees quit in May. We nearly kept up with incoming
 volumne in the following two months, but fell behind with the
 increase in Aug and havent been able to catch up since''')

ax3.spines['bottom'].set_color('grey')
ax3.spines['left'].set_color('grey')

ax3.tick_params(bottom = False, color = "grey") 
```
