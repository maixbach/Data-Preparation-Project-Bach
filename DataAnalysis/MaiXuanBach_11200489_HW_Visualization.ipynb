## 2. Product_price

```python
df = pd.read_excel("./Data Visualize HW/Product_Price.xlsx")
df
```

```python
df1 = df.T
df1.columns = df["Unnamed: 0"]
df1.drop(index = ["Unnamed: 0", "Unnamed: 1"], inplace = True)
df1.reset_index(inplace= True)
#df1.rename({"index" : "Year"}, inplace= True)
df1
```

```python
fig, ax = plt.subplots(figsize = (10, 4))

ax.spines[["top", "right"]].set_visible(False) 

ax.set_title('Average Retail Product Price per Year', fontweight="bold", size = 14)

# Align the minor tick label
for label in ax.get_xticklabels():
    label.set_horizontalalignment('right')

ax.plot(df1["index"], df1["Product A"], color = "#204D7D", label ='Product A', linewidth=3.0) #blue
ax.plot(df1["index"], df1["Product B"], color = "#900C3F", label ='Product B', linewidth=3.0) # red
ax.plot(df1["index"], df1["Product C"], color = "#527D4C", label ='Product C', linewidth=3.0) # green
ax.plot(df1["index"], df1["Product D"], color = "#5E285F", label ='Product D', linewidth=3.0) #purple
ax.plot(df1["index"], df1["Product E"], color = "#1A5050", label ='Product E', linewidth=3.0)

ax.grid(visible = True, axis = "y")
ax.set_yticks([i for i in range(0, 600, 100)])
ax.set_yticklabels([ "$" + x.astype("str") for x in ax.get_yticks()]) 




ax.legend(loc='lower center', bbox_to_anchor=(0.5, -0.25),prop={'size': 12},
          fancybox=True, shadow=True, frameon=False, ncol=5)

ax.spines['bottom'].set_color('grey')
ax.spines['left'].set_color('grey')
```
